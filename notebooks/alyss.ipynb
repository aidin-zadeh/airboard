```python
# import dependencies
import os, inspect
import pandas as pd
import numpy as np

# from glob import glob, iglob
```

```python
# get current dir
curr_dir = os.path.dirname(inspect.getabsfile(inspect.currentframe()))
root_dir = os.path.dirname(curr_dir)

print(curr_dir)
print(root_dir)

data_dir = os.path.join(root_dir, "airboard", "data")
raw_data_dir = os.path.join(data_dir, "raw")

print(data_dir)
print(raw_data_dir)
```

```python
# def make_mongo_db():

# read in each csv
data = pd.DataFrame()
with os.scandir(raw_data_dir) as dir:
    for f in dir:
        if f.name.endswith('CARRIER.zip'):
            print(f"Importing {f.name}")
            fpath = os.path.join(raw_data_dir, f.name)
            df = pd.read_csv(fpath, usecols=['PASSENGERS', 'FREIGHT', 'MAIL', 
                                             'DISTANCE', 'UNIQUE_CARRIER_NAME', 
                                             'ORIGIN_CITY_MARKET_ID', 'ORIGIN', 
                                             'ORIGIN_CITY_NAME', 'ORIGIN_STATE_ABR', 
                                             'DEST_CITY_MARKET_ID', 'DEST', 
                                             'DEST_CITY_NAME', 'DEST_STATE_ABR', 
                                             'YEAR', 'MONTH', 'CLASS'])
            df.head(1)
            
            data = data.append(df, ignore_index=True)          
print("Import complete.")
# data.head()

# Clean data/drop NaNs
print(f"original --> {len(data.columns.tolist())} columns, {len(data)} rows")
data.dropna(inplace=True)
print(f"cleaned --> {len(data.columns.tolist())} columns, {len(data)} rows")

# usecols=['PASSENGERS', 'FREIGHT', 'MAIL', 'DISTANCE', 'UNIQUE_CARRIER_NAME', 
#          'ORIGIN_CITY_MARKET_ID', 'ORIGIN', 'ORIGIN_CITY_NAME', 'ORIGIN_STATE_ABR', 
#          'DEST_CITY_MARKET_ID', 'DEST', 'DEST_CITY_NAME', 'DEST_STATE_ABR', 
#          'YEAR', 'MONTH', 'CLASS']

print(data.columns.tolist())
data.head()
```

-------
### Some identification data is redundant and can be removed to save space in a master data file

airports.csv can be converted to dict for conversion between ID types/grouping categories 

```python
# Clean redundant data
data['ORIGIN_CITY_NAME'] = data['ORIGIN_CITY_NAME'].apply(lambda x: x.split(',')[0])
data['DEST_CITY_NAME'] = data['DEST_CITY_NAME'].apply(lambda x: x.split(',')[0])

data.head()
```

```python
cols = ['AIRPORT', 'CITY', 'STATE', 'CITY_MARKET_ID']

origins = data[['ORIGIN', 'ORIGIN_CITY_NAME', 'ORIGIN_STATE_ABR', 'ORIGIN_CITY_MARKET_ID']]
origins = origins.drop_duplicates()
origins.columns = cols

dests = data[['DEST', 'DEST_CITY_NAME', 'DEST_STATE_ABR', 'DEST_CITY_MARKET_ID']]
dests = dests.drop_duplicates()
dests.columns = cols

airports = dests.merge(origins, how='left')
print(len(origins), len(dests), len(airports))
airports.head()
```

```python
airports.to_csv(os.path.join(data_dir, 'int', 'airport_info.csv'))
print('done.')
```

#### Function for getting top 10 groupby lists

```python
def top10(groupByList):
    agg = flights.groupby(groupByList).size()
    return agg.nlargest(10)
```
